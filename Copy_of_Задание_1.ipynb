##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
import numpy as np
import csv
```

```python
#Агрегация
with open("Global YouTube Statistics.csv", encoding= 'ISO-8859-9') as f:
  arr = list(csv.reader(f, delimiter=','))

arr = np.array(arr)
arr
```

```python
category = arr[1:,4]
subscribers = arr[1:,2]
#print(category)
reg = {}
for i in range (len(category)):
  if category[i] not in reg.keys():
    reg.setdefault(category[i], [])
  reg[category[i]].append(subscribers[i])

print(reg)
```

```python
for k in reg.keys():
  for i in range (len(reg[k])):
    reg[k][i] = int(reg[k][i])
  print(k, "\n", "Среднее :", np.mean(reg[k]), '\n', "Медиана :", np.median(reg[k]))

```
