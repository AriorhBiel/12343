````python
"""

Ваш код здесь

"""
from statistics import median, mean

class Outline:
  def __init__(self, df):
    self.df = df

  def delete_outline_by_mean(self, column_name):

    error = []

    for item in self.df[column_name]:
      arr = list(self.df[column_name])
      X = arr.pop(item)
      print("1",len(arr), item, arr)

      srednee = np.mean(arr)
      std = np.std(arr)

      Left = srednee - std*3
      Right = srednee + std*3


      if item < Left or item > Right:
        error.append(item)

    return error
````
